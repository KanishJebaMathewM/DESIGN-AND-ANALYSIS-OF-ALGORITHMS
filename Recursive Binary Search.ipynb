```python
import time
import matplotlib.pyplot as plt
import random
```

```python
def binary_search_recursive(arr, target, low, high):
    if low > high:
        return -1  # Element not found

    mid = (low + high) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] > target:
        return binary_search_recursive(arr, target, low, mid - 1)
    else:
        return binary_search_recursive(arr, target, mid + 1, high)
```

```python
def measure_time(n):
    arr = list(range(n))  # Sorted array
    target = random.randint(0, n - 1)  # Random element

    start_time = time.time()
    binary_search_recursive(arr, target, 0, n - 1)
    end_time = time.time()

    return (end_time - start_time) * 1e6  # time in microseconds
```

```python
ns = [10, 100, 500, 1000, 5000, 10000, 20000, 50000, 100000]
times = []

for n in ns:
    t = measure_time(n)
    times.append(t)

# Optional: Display values
for i in range(len(ns)):
    print(f"n = {ns[i]}, Time = {times[i]:.4f} μs")
```

```python
plt.plot(ns, times, marker='o', color='blue')
plt.title('Time Taken vs Number of Elements (Recursive Binary Search)')
plt.xlabel('Number of Elements (n)')
plt.ylabel('Time Taken (μs)')
plt.grid(True)
plt.show()
```
