```python
import random
import time
import matplotlib.pyplot as plt

# Linear Search Function
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1

# Generate list of random integers
n = 10000
arr = [random.randint(0, 100000) for _ in range(n)]

# Insert known values for controlled testing
arr[0] = -1                # Best case: search -1
arr[n // 2] = -2           # Average case: search -2
# Do not insert -3: for worst case, -3 is not in list

# Timing Best Case
start_best = time.time()
linear_search(arr, -1)
end_best = time.time()
time_best = (end_best - start_best) * 1000  # in milliseconds

# Timing Average Case
start_avg = time.time()
linear_search(arr, -2)
end_avg = time.time()
time_avg = (end_avg - start_avg) * 1000

# Timing Worst Case
start_worst = time.time()
linear_search(arr, -3)
end_worst = time.time()
time_worst = (end_worst - start_worst) * 1000

# Display Times
print(f"Best Case Time: {time_best:.5f} ms")
print(f"Average Case Time: {time_avg:.5f} ms")
print(f"Worst Case Time: {time_worst:.5f} ms")

# Plotting
cases = ['Best Case', 'Average Case', 'Worst Case']
times = [time_best, time_avg, time_worst]

plt.figure(figsize=(8, 5))
plt.bar(cases, times, color=['green', 'orange', 'red'])
plt.title('Linear Search Performance')
plt.ylabel('Time (milliseconds)')
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

# Linear Search with comparison count
def linear_search_with_comparisons(arr, key):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == key:
            return i, comparisons
    return -1, comparisons

# Input sizes
input_sizes = [100, 500, 1000, 5000, 10000]

# Results
best_comparisons = []
avg_comparisons = []
worst_comparisons = []

# Test each input size
for n in input_sizes:
    arr = [random.randint(0, 100000) for _ in range(n)]
    
    arr[0] = -1           # Best case: key at index 0
    arr[n // 2] = -2      # Average case: key at middle
    # -3 is not added for worst case

    # Best Case
    _, best = linear_search_with_comparisons(arr, -1)
    # Average Case
    _, avg = linear_search_with_comparisons(arr, -2)
    # Worst Case
    _, worst = linear_search_with_comparisons(arr, -3)

    best_comparisons.append(best)
    avg_comparisons.append(avg)
    worst_comparisons.append(worst)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(input_sizes, best_comparisons, marker='o', label='Best Case (O(1))')
plt.plot(input_sizes, avg_comparisons, marker='o', label='Average Case (O(n/2))')
plt.plot(input_sizes, worst_comparisons, marker='o', label='Worst Case (O(n))')
plt.title('Linear Search: Comparisons vs Input Size')
plt.xlabel('Input Size (n)')
plt.ylabel('Number of Comparisons')
plt.legend()
plt.grid(True)
plt.show()

# Print Table
print("Input Size | Best Case | Average Case | Worst Case")
for i in range(len(input_sizes)):
    print(f"{input_sizes[i]:>10} | {best_comparisons[i]:>9} | {avg_comparisons[i]:>13} | {worst_comparisons[i]:>10}")
```
